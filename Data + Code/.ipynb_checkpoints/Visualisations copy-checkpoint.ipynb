```python
import csv
import pandas as pd
import math
import numpy as np
from ast import literal_eval
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
pno = 3

fout=open("test1-all.csv","w")
# first file:
for line in open("Test1/test1-ID1-P1.csv"):
    fout.write(line)
# now the rest:    
for num in range(2,pno):
    f = open("Test1/test1-ID"+str(num)+"-P1.csv")
    f.__next__() # skip the header
    for line in f:
         fout.write(line)
    f.close() # not really needed
fout.close()
```

```python
target1 = pd.read_csv("test1-all.csv")

target1 = target1.loc[(target1['targetX'] == 100) & (target1['targetY'] == 50)]
```

```python
def calculateDistance(x1,y1,x2,y2):
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)  
    return dist


def calculate(x1,y1):
    return x1-y1
```

```python
distance = target1[['targetX', 'targetY', 'mouseX', 'mouseY']].copy()
target1['distance'] = target1.apply(lambda x: calculateDistance(x.targetx, x.targety, x.mousex, x.mousey), axis=1)
target1['distanceStart'] = target1.apply(lambda x: calculateDistance(650, 375, x.mousex, x.mousey), axis=1)
target1['velocity'] = target1['distanceStart']/target1['testTime']
target1['acceleration'] = (target1['velocity'] - target1['velocity'].shift(1))/target1['testTime']
target1
```

```python
fig = plt.figure()
for i in range (1,pno):
    x = target1.loc[(target1['participant'] == i)]
    plt.plot(x['testTime'], x['distance'])


plt.xlabel('Time (seconds)')
plt.ylabel('Position')
```

```python
fig = plt.figure()
for i in range (1,pno):
    x = target1.loc[(target1['participant'] == i)]
    plt.plot(x['testTime'], x['distanceStart'])

fig.suptitle('Position Relative to Start', fontsize=20)
plt.xlabel('Time (seconds)')
plt.ylabel('Position')
```

```python
fig = plt.figure()
for i in range (1,pno):
    x = target1.loc[(target1['participant'] == i)]
    plt.plot(x['distanceStart'], x['velocity'])

fig.suptitle('Phasespace Plot', fontsize=20)
plt.xlabel('Position')
plt.ylabel('Velocity')

#make target 0
```

```python
fig = plt.figure()
for i in range (1,2):
    x = target1.loc[(target1['participant'] == i)]
    plt.plot(x['distanceStart'], x['acceleration'])

fig.suptitle('Hooke Plot', fontsize=20)
plt.xlabel('Position')
plt.ylabel('Acceleration')

```
